(a) Let $ X =\begin {pmatrix}x\\ y \end{pmatrix} \implies âˆ‡f(X) = \begin{pmatrix} \frac{\pi}{2}y^2\cdot \sin\frac{\pi x}{2}\\ 2(1-cos\frac{\pi x}{2})y\end{pmatrix}$. 

Let $\nabla f = 0 \implies$ 
\begin{align}
        \left\{
        \begin{array}{cl}
        \frac{\pi}{2}y^2\cdot \sin\frac{\pi x}{2}=0 \\
         2(1-cos\frac{\pi x}{2})y= 0.
        \end{array}
        \right.
    \end{align}

$\implies$Set of minimizer $M =\{(x,0)|x\in[-1,1]\} \cup \{(0,y)|y\in[-1,1]\} $ 


(b)The code and plot of cost function is as follows:

```python
import matplotlib.pyplot as plt
from math import cos, pi, sin
import numpy as np

#define the cost function
def f(x,y):
    return (1-cos(pi*x/2))*y**2

xi, yi = np.meshgrid( np.linspace(-1, 1, 1000), np.linspace(-1, 1, 1000) )
#evalute the function
f_eval = np.zeros(xi.shape)
mask = []
for i in range(1000):
    for j in range(1000):
        f_eval[i, j] = f(xi[i,j], yi[i,j])
#plot
plt.figure(figsize=(12,10))
plt.axis('equal')
plt.contourf(xi, yi, f_eval, levels =10)
plt.title("Cost function, level sets")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()  
plt.show()
```

(c)The trajectories are plotted as follows with 100 randomly chosen $r_0 \in B$

```python
import colorcet as cc
#gradient function
def grad_f(xi, yi):
    return np.array([ pi/2 * yi**2 *sin(pi*xi/2), 2*(1-cos(pi/2*xi))*yi])
plt.figure(figsize=(10,10))
plt.axis('equal') 
im = plt.contourf(xi, yi, f_eval, levels = 10, cmap=cc.cm.bmy)
plt.title("Cost function, 10 steps gradient descent")
for i in range(100):
    p = [tuple(np.random.uniform(-1, 1, 2))]
    for j in range(10):
        p.append(tuple(np.subtract(p[-1], grad_f(*p[-1]))))
    p = np.array(p)
    plt.plot(*p.T, c='k', linewidth=0.5, markersize=1, marker = '.')
```
